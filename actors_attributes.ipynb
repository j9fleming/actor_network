```python
import pandas as pd
```

```python
names_df = pd.read_csv('~/soc_final/actors_names.tsv', sep = '\t')
```

```python
titles_df = pd.read_csv('~/soc_final/titles.tsv', sep = '\t')
```

```python
top_actors_df = pd.read_csv("~/top_grossing_actors.csv")
```

```python
#actors that have titles they are known for
actors_df = names_df[(names_df['primaryProfession'].str.contains('actor') |
                      names_df['primaryProfession'].str.contains('actress')) &
                      names_df['knownForTitles'].str.contains('tt')].reset_index(drop = True)
```

```python
#filtering imdb actors to top 300 actors
top_actors = top_actors_df['Name'].tolist()
actors_df = actors_df[actors_df['primaryName'].isin(top_actors)]
```

```python
#filtering out duplicates based on birthyear
actors_df = actors_df[actors_df['birthYear'] != '\\N']
```

```python
#getting rest of the duplicates
name_duplicates_df = actors_df[actors_df.duplicated('primaryName')]['primaryName'].unique().tolist()
```

```python
#filtered names
fnd = ['nm0000148', #Harrison Ford
       'nm0262635', #Chris Evans
       'nm0268199', #Colin Farrell
       'nm0362766', #Tom Hardy
       'nm0748620', #Paul Rudd
       'nm0788335', #Michael Shannon
       'nm0000226' #Will Smith
                 ]

fnd_df = actors_df[actors_df['nconst'].isin(fnd)]

```

```python
#Remove all records with duplicates
actors_df = actors_df.drop_duplicates('primaryName',keep = False)
#combine actors_df with fnd (duplicate actors that were just removed)
if(len(actors_df['primaryName']) != 142):
    actors_df = pd.concat([actors_df, fnd_df]).reset_index(drop = True)

test = actors_df['primaryName'].tolist()
print(len(test))
```

```python
if(len(actors_df['primaryName']) != 299):
    actors_df = pd.concat([actors_df, missing_names_df]).reset_index(drop = True)
    actors_df['knownForTitles'] = actors_df['knownForTitles'].apply(lambda x: x.split(',') )
actors_df
```

```python
titles_set = set()
for ind in actors_df.index:
    kt = actors_df['knownForTitles'][ind]
    for t in kt:
        if t not in titles_set:
            titles_set.add(t)
len(titles_set)
```

```python
titles_df = titles_df[titles_df['tconst'].isin(titles_set)]
```
